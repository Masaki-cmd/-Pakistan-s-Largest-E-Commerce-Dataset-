## データ理解レポート (Data Understanding Report)

```python
import pandas as pd
import os
import glob
import numpy as np # For potential random seed or statistical operations if needed later

# 乱数シードの固定（必要時のみ。今回はEDAが決定的なので不要）
# np.random.seed(42)
```

```python
def perform_data_understanding(df: pd.DataFrame, file_name: str):
    """
    DataFrameのデータ理解（Data Understanding）を実行し、結果を日本語で出力する。

    Args:
        df (pd.DataFrame): データ理解の対象となるDataFrame。
        file_name (str): 処理中のファイル名。
    """
    print(f"--- ファイル: {file_name} ---")

    # 1. 行数・列数
    num_rows, num_cols = df.shape
    print(f"\n1. 行数・列数: (行: {num_rows}, 列: {num_cols})")

    # 2. データ型一覧
    print("\n2. データ型一覧:")
    print(df.dtypes.to_string())

    # 3. 欠損率（列別）
    missing_percentage = df.isnull().sum() / len(df) * 100
    print("\n3. 欠損率（列別）:")
    print(missing_percentage[missing_percentage > 0].sort_values(ascending=False).to_string())
    if missing_percentage.max() == 0:
        print("  欠損値はありません。")

    # 4. 重複行数
    duplicate_rows = df.duplicated().sum()
    print(f"\n4. 重複行数: {duplicate_rows}")

    # 5. 主キー候補評価（理由付き）
    print("\n5. 主キー候補評価:")
    pk_candidates = []
    for col in df.columns:
        if df[col].is_unique:
            pk_candidates.append(f"  - '{col}': 全ての値が一意であるため、単一列の主キー候補です。")

    if not pk_candidates:
        # 複合主キーの簡易的な評価（全列の組み合わせがユニークであるか）
        if df.drop_duplicates().shape[0] == df.shape[0]:
             pk_candidates.append("  - 複合主キー候補: 全ての列の組み合わせが一意であるため、複合主キー候補です。特定の単一列では主キーとなりません。")
        else:
            pk_candidates.append("  - 主キー候補なし: 全ての行の組み合わせも一意ではないか、または特定の単一列で主キー候補となるものはありませんでした。データ内容を詳細に確認する必要があります。")

    for cand in pk_candidates:
        print(cand)

    # 6. 分析可能な粒度（1行定義）
    print("\n6. 分析可能な粒度（1行定義）:")
    # ファイル名に基づいて粒度を推測
    granularity = "データセットの内容を詳細に確認しないと定義できません。"
    if 'california_housing_train' in file_name or 'california_housing_test' in file_name:
        granularity = "カリフォルニア州の特定の地域（ブロック）における住宅に関する統計データの一行です。"
    elif 'mnist_train_small' in file_name or 'mnist_test' in file_name:
        granularity = "手書き数字の画像データ（各ピクセルの値とラベル）の一行です。"
    elif 'anscombe' in file_name:
        granularity = "アンコムの例示データセットの特定のグループにおけるXとYの値の一行です。"
    else:
        granularity = "一般的なCSVファイルのため、具体的な列名を見ないと特定できません。"
    print(f"  - {granularity}")

    print("\n" + "="*60 + "\n")
```

```python
# CSVファイルが格納されているディレクトリ
data_dir = '/content/sample_data/'

# ディレクトリ内のすべてのCSVファイルを取得
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))

if not csv_files:
    print(f"エラー: '{data_dir}' ディレクトリにCSVファイルが見つかりませんでした。")
else:
    print(f"以下のCSVファイルが見つかりました: {[os.path.basename(f) for f in csv_files]}\n")
    for csv_file_path in csv_files:
        try:
            df = pd.read_csv(csv_file_path)
            perform_data_understanding(df, os.path.basename(csv_file_path))
        except Exception as e:
            print(f"ファイル '{os.path.basename(csv_file_path)}' の読み込みまたは処理中にエラーが発生しました: {e}")
```

## データクレンジングレポート (Data Cleaning Report)

```python
def perform_data_cleaning(df: pd.DataFrame, file_name: str) -> dict:
    """
    DataFrameのデータクレンジングを実行し、結果を日本語で出力する。

    Args:
        df (pd.DataFrame): データクレンジングの対象となるDataFrame。
        file_name (str): 処理中のファイル名。

    Returns:
        dict: クリーニング結果のサマリー。
    """
    print(f"--- ファイル: {file_name} のデータクレンジング ---\n")
    cleaning_report = {
        "file_name": file_name,
        "original_rows": len(df),
        "changes": [],
        "affected_rows": 0,
        "business_impact": [],
        "data_quality_score": 100
    }
    initial_issue_rows = set() # 一意な問題のある行インデックスを収集

    # 0. 欠損処理 (Data Understandingの結果から、今回は欠損値はなし)
    # 1. 型変換 (Data Understandingの結果から、現状のデータタイプが適切と判断されたため、特になし)

    # 2. 異常値チェック（業務ロジック） & 外れ値検証
    # MNISTデータセットの処理
    if 'mnist' in file_name:
        label_col = 'label' # 列0を'label'にリネームしたことを想定
        pixel_cols = [col for col in df.columns if col.startswith('pixel_')]

        # ラベル列の範囲外チェック (0-9)
        label_anomalies = df[(df[label_col] < 0) | (df[label_col] > 9)]
        if not label_anomalies.empty:
            cleaning_report["changes"].append(f"ラベル列 '{label_col}' に範囲外の異常値（0-9以外）を {len(label_anomalies)} 件検出しました。")
            cleaning_report["business_impact"].append("モデル学習の正確性への影響の可能性。")
            initial_issue_rows.update(label_anomalies.index.tolist())

        # ピクセル値の範囲外チェック (0-255)
        pixel_out_of_range_indices = set()
        for col in pixel_cols:
            anomalies = df[(df[col] < 0) | (df[col] > 255)]
            if not anomalies.empty:
                pixel_out_of_range_indices.update(anomalies.index.tolist())

        if pixel_out_of_range_indices:
            cleaning_report["changes"].append(f"ピクセル値に範囲外の異常値（0-255以外）を {len(pixel_out_of_range_indices)} 件検出しました。")
            cleaning_report["business_impact"].append("画像データの表現品質への影響、モデル学習の正確性への影響の可能性。")
            initial_issue_rows.update(pixel_out_of_range_indices)

        if not initial_issue_rows:
            cleaning_report["changes"].append("MNISTデータセットにおける異常値・外れ値は検出されませんでした。")


    # California Housingデータセットの処理
    elif 'california_housing' in file_name:
        numeric_cols = df.select_dtypes(include=['float64']).columns

        # 外れ値検出 (IQR法)
        for col in numeric_cols:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            if not outliers.empty:
                cleaning_report["changes"].append(f"列 '{col}' にIQR法による外れ値（{len(outliers)} 件）を検出しました。")
                cleaning_report["business_impact"].append(f"'{col}' の外れ値がモデルの頑健性や統計的分析に影響を与える可能性があります。")
                initial_issue_rows.update(outliers.index.tolist())

        # 異常値チェック (業務ロジック)

        # 経度・緯度の範囲チェック (おおよそのカリフォルニア州の範囲)
        lon_anomalies = df[(df['longitude'] < -125) | (df['longitude'] > -113)]
        lat_anomalies = df[(df['latitude'] < 32) | (df['latitude'] > 42)]
        if not lon_anomalies.empty:
            cleaning_report["changes"].append(f"列 'longitude' にカリフォルニア州の範囲外の異常値（{len(lon_anomalies)} 件）を検出しました。")
            cleaning_report["business_impact"].append("地理空間分析の正確性に影響を与える可能性があります。")
            initial_issue_rows.update(lon_anomalies.index.tolist())
        if not lat_anomalies.empty:
            cleaning_report["changes"].append(f"列 'latitude' にカリフォルニア州の範囲外の異常値（{len(lat_anomalies)} 件）を検出しました。")
            cleaning_report["business_impact"].append("地理空間分析の正確性に影響を与える可能性があります。")
            initial_issue_rows.update(lat_anomalies.index.tolist())

        # 各数値列の非負チェック
        for col in ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']:
            negative_anomalies = df[df[col] < 0]
            if not negative_anomalies.empty:
                cleaning_report["changes"].append(f"列 '{col}' に負の値の異常値（{len(negative_anomalies)} 件）を検出しました。")
                cleaning_report["business_impact"].append(f"'{col}' の負の値が統計指標の誤解釈やモデルの性能低下につながる可能性があります。")
                initial_issue_rows.update(negative_anomalies.index.tolist())

        # total_bedrooms > total_rooms のチェック
        bedroom_room_anomalies = df[df['total_bedrooms'] > df['total_rooms']]
        if not bedroom_room_anomalies.empty:
            cleaning_report["changes"].append(f"列 'total_bedrooms' が 'total_rooms' を超える異常値（{len(bedroom_room_anomalies)} 件）を検出しました。")
            cleaning_report["business_impact"].append("住宅特性に関する指標の整合性問題や、派生特徴量の計算誤りの可能性。")
            initial_issue_rows.update(bedroom_room_anomalies.index.tolist())

        # households > population のチェック
        household_pop_anomalies = df[df['households'] > df['population']]
        if not household_pop_anomalies.empty:
            cleaning_report["changes"].append(f"列 'households' が 'population' を超える異常値（{len(household_pop_anomalies)} 件）を検出しました。")
            cleaning_report["business_impact"].append("人口統計学的分析や世帯関連指標の整合性問題。")
            initial_issue_rows.update(household_pop_anomalies.index.tolist())

        if not initial_issue_rows:
            cleaning_report["changes"].append("California Housingデータセットにおけるビジネスロジックに基づく異常値・外れ値は検出されませんでした。")


    # 問題のある一意な行数を更新
    cleaning_report["affected_rows"] = len(initial_issue_rows)

    # データ品質スコアの計算
    if cleaning_report["original_rows"] > 0:
        issue_percentage = (cleaning_report["affected_rows"] / cleaning_report["original_rows"]) * 100
        cleaning_report["data_quality_score"] = max(0, 100 - issue_percentage) # スコアは0未満にならないように調整
    else:
        cleaning_report["data_quality_score"] = 0 # 空のDataFrameはスコア0とする

    print("### クリーニング結果サマリー ###")
    print(f"**変更内容**: {'、'.join(cleaning_report['changes']) if cleaning_report['changes'] else '特になし。'}")
    print(f"**影響件数（一意な行）**: {cleaning_report['affected_rows']} 件")
    print(f"**ビジネス影響**: {'、'.join(cleaning_report['business_impact']) if cleaning_report['business_impact'] else '特になし。'}")
    print(f"**データ品質スコア**: {cleaning_report['data_quality_score']:.2f}/100")
    print("\n" + "="*60 + "\n")

    return cleaning_report
```

```python
# CSVファイルが格納されているディレクトリ
data_dir = '/content/sample_data/'

# ディレクトリ内のすべてのCSVファイルを取得
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))

if not csv_files:
    print(f"エラー: '{data_dir}' ディレクトリにCSVファイルが見つかりませんでした。")
else:
    print(f"以下のCSVファイルが見つかりました: {[os.path.basename(f) for f in csv_files]}\n")
    for csv_file_path in csv_files:
        file_name = os.path.basename(csv_file_path)
        try:
            if 'mnist' in file_name:
                # MNISTデータはヘッダーなしで、最初の列がラベル
                df = pd.read_csv(csv_file_path, header=None)
                df.rename(columns={0: 'label'}, inplace=True)
                pixel_col_names = [f'pixel_{i}' for i in range(df.shape[1] - 1)]
                df.columns = ['label'] + pixel_col_names.copy()
            else: # california_housing files
                df = pd.read_csv(csv_file_path)

            perform_data_cleaning(df, file_name)
        except Exception as e:
            print(f"ファイル '{file_name}' の読み込みまたは処理中にエラーが発生しました: {e}")
```

## 探索的データ分析レポート (Exploratory Data Analysis Report)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import pearsonr

def print_eda_section_header(title):
    print(f"\n### {title} ###")
    print("-" * (len(title) + 6))

def print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis):
    print(f"  **目的**: {purpose}")
    print(f"  **方法**: {method}")
    print(f"  **結果（数値）**: {result}")
    print(f"  **解釈**: {interpretation}")
    print(f"  **ビジネス意味**: {business_impact}")
    print(f"  **リスク**: {risk}")
    print(f"  **意思決定影響度**: {decision_impact}")
    print(f"  **仮説**: {hypothesis}")
    print("\n")

def perform_mnist_eda(df: pd.DataFrame, file_name: str):
    print(f"--- ファイル: {file_name} のEDA ---\n")
    unnecessary_analyses = []

    # Distribution Analysis: Label Distribution
    print_eda_section_header("分布分析: ラベル（数字）の分布")
    label_counts = df['label'].value_counts().sort_index()
    purpose = "各手書き数字（ラベル）の出現頻度を理解する。"
    method = "'label'列のカウントと比率を計算。"
    result = f"ラベルごとの件数:\n{label_counts.to_string()}\n総件数: {len(df)}"
    interpretation = f"最も多い数字は{label_counts.idxmax()}で{label_counts.max()}件、最も少ない数字は{label_counts.idxmin()}で{label_counts.min()}件です。データセット内の数字の分布は比較的均一または特定の数字に偏りが見られます。"
    business_impact = "もし分布に大きな偏りがある場合、モデルが少数派の数字を識別しにくくなる可能性があるため、学習データのバランス調整を検討する必要がある。"
    risk = "不均衡データによるモデルの汎化性能低下。"
    decision_impact = "High"
    hypothesis = "ラベルの分布に偏りがある場合、モデルの精度が低いラベルに対しては改善策が必要である。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Distribution Analysis: Pixel Value Statistics (overall)
    print_eda_section_header("分布分析: ピクセル値の統計概要")
    pixel_cols = [col for col in df.columns if col.startswith('pixel_')]
    pixel_stats = df[pixel_cols].agg(['mean', 'std', 'min', 'max']).T
    purpose = "ピクセル値の全体的な特性（明るさ、コントラスト）を把握する。"
    method = "全ピクセル列の平均、標準偏差、最小値、最大値を計算。"
    result = f"ピクセル値の統計:\n{pixel_stats.to_string()}"
    interpretation = "ピクセル値は0から255の範囲にあり、平均値や標準偏差から画像の明るさや多様性を概観できます。"
    business_impact = "特定のピクセルや画像全体に異常な値がないか確認し、画像前処理の必要性を判断する。"
    risk = "極端なピクセル値がモデルの学習を阻害する可能性。"
    decision_impact = "Medium"
    hypothesis = "ピクセル値の範囲外の異常値はデータクレンジングで除去されており、全体の統計値は妥当な範囲内にあるため、追加の処置は不要である。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Group Comparison: Mean Pixel Value per Label (a simple example)
    print_eda_section_header("比較分析: ラベルごとの平均ピクセル値")
    mean_pixel_per_label = df.groupby('label')[pixel_cols].mean().mean(axis=1)
    purpose = "各数字が全体としてどれくらいの明るさで書かれているかを比較する。"
    method = "ラベルごとにピクセル値の平均を算出し、それをさらに平均して比較。"
    result = f"ラベルごとの平均ピクセル値（全体の平均）:\n{mean_pixel_per_label.to_string()}"
    interpretation = "数字によって平均的な明るさ（ピクセルが埋まっている度合い）が異なることがわかります。例えば、'1'は細いため平均が低く、'8'は複雑なため平均が高い傾向があるかもしれません。"
    business_impact = "数字の形状特性に基づくモデルの識別ロジックを理解する手助けとなる。"
    risk = "特定の数字が極端に薄い・濃い場合、認識率に影響する可能性。"
    decision_impact = "Low"
    hypothesis = "平均ピクセル値が低い（薄い）数字は、誤認識率が高くなる傾向がある。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Contribution/Pareto: Pixels with Highest Variance
    print_eda_section_header("構造分析: 分散の大きいピクセル（寄与度）")
    pixel_variances = df[pixel_cols].var().sort_values(ascending=False)
    top_10_variance_pixels = pixel_variances.head(10)
    purpose = "手書き数字の識別に最も寄与する可能性のあるピクセル（分散が大きいピクセル）を特定する。"
    method = "全ピクセル列の分散を計算し、上位10件を抽出。"
    result = f"分散上位10ピクセル:\n{top_10_variance_pixels.to_string()}"
    interpretation = "分散が大きいピクセルは、数字が書かれる領域の中心付近や、数字の形状が大きく変化する境界部分に位置している可能性があり、数字の識別に重要な情報を含んでいます。"
    business_impact = "特徴量選択や次元削減の際に、より重要なピクセルを優先するヒントとなる。"
    risk = "分散が大きくてもノイズである可能性も排除できない。"
    decision_impact = "Medium"
    hypothesis = "分散の大きいピクセルは、数字の分類精度に対して高い寄与度を持つ。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Trend Analysis (Not Applicable for MNIST)
    print_eda_section_header("時系列分析: (適用外)")
    unnecessary_analyses.append("MNISTデータは時系列データではないため、トレンド分析は今回の意思決定に不要です。")

    print(f"\n**今回の意思決定に不要だった分析:**\n{'\n'.join(unnecessary_analyses) if unnecessary_analyses else '特になし。'}")
    print("\n" + "="*60 + "\n")

def perform_california_housing_eda(df: pd.DataFrame, file_name: str):
    print(f"--- ファイル: {file_name} のEDA ---\n")
    unnecessary_analyses = []

    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    # 目的変数を中心にEDAを進める
    target_col = 'median_house_value'
    if target_col not in numeric_cols:
        print(f"警告: 目的変数 '{target_col}' がデータフレームに存在しません。EDAを中断します。")
        return

    feature_cols = [col for col in numeric_cols if col != target_col]

    # Distribution Analysis: Target Variable
    print_eda_section_header(f"分布分析: {target_col} の分布")
    purpose = f"{target_col} の中央値、平均、歪度、尖度を理解する。"
    method = f"{target_col} の記述統計量（平均、中央値、標準偏差、歪度、尖度）を計算。"
    desc_stats = df[target_col].describe()
    skewness = df[target_col].skew()
    kurtosis = df[target_col].kurt()
    result = f"記述統計量:\n{desc_stats.to_string()}\n歪度: {skewness:.2f}\n尖度: {kurtosis:.2f}"
    interpretation = f"{target_col} は平均値と中央値に差が見られ、分布の歪度や尖度から正規分布からの乖離が示唆されます。特に、歪度が正の値であることから、高価格帯に少数の住宅が存在する（右に裾が長い）ことが考えられます。"
    business_impact = "住宅価格予測モデルを構築する際、目的変数の分布特性を考慮したモデル選択や変換（例：対数変換）が必要となる可能性を示唆。"
    risk = "目的変数が正規分布から大きく外れる場合、線形モデルの性能が低下する可能性。"
    decision_impact = "High"
    hypothesis = "`median_house_value`は正に歪んだ分布を持つため、予測モデルの性能向上のためにターゲット変数の対数変換を検討すべきである。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Distribution Analysis: Key Features (e.g., median_income, housing_median_age)
    print_eda_section_header("分布分析: 主要な特徴量の分布")
    for col in ['median_income', 'housing_median_age']:
        if col in df.columns:
            desc_stats = df[col].describe()
            skewness = df[col].skew()
            kurtosis = df[col].kurt()
            purpose = f"'{col}' の分布特性を理解し、データ前処理の必要性を評価する。"
            method = f"'{col}' の記述統計量（平均、中央値、標準偏差、歪度、尖度）を計算。"
            result = f"'{col}' の記述統計量:\n{desc_stats.to_string()}\n歪度: {skewness:.2f}\n尖度: {kurtosis:.2f}"
            interpretation = f"'{col}' もまた正規分布から乖離している可能性があり、特に 'median_income' は高所得世帯に少数の値が存在する（右に裾が長い）可能性があります。"
            business_impact = "特徴量エンジニアリングやスケーリングの必要性を判断する。外れ値の影響を考慮する必要がある。"
            risk = "外れ値や歪みがモデルの学習を不安定にする可能性。"
            decision_impact = "Medium"
            hypothesis = f"'{col}' の歪みはモデルの性能に影響を与えるため、適切な変換（例：ログ変換や標準化）が必要である。"
            print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Group Comparison: Median House Value by Housing Median Age Bins
    print_eda_section_header("比較分析: 築年数区分別の中央住宅価格")
    df['age_bin'] = pd.cut(df['housing_median_age'], bins=[0, 10, 20, 30, 40, 50, np.inf], labels=['0-10', '11-20', '21-30', '31-40', '41-50', '50+'], right=False)
    grouped_age_value = df.groupby('age_bin')[target_col].median().fillna(0)
    purpose = "築年数が中央住宅価格に与える影響を概観する。"
    method = "'housing_median_age' を区分し、各区分の中央住宅価格の中央値を比較。"
    result = f"築年数区分別の中央住宅価格:\n{grouped_age_value.to_string()}"
    interpretation = "築年数が増加するにつれて、中央住宅価格が特定のパターンを示す傾向があるかもしれません。例えば、築浅と築古で価格帯が異なることが考えられます。"
    business_impact = "不動産開発や投資戦略において、ターゲットとなる築年数帯を特定する上で有用。"
    risk = "他の要因（地域、所得など）が価格に大きく影響している可能性があり、築年数単体での解釈は限定的。"
    decision_impact = "Medium"
    hypothesis = "築年数が若いほど、中央住宅価格は高い傾向がある。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)
    df.drop('age_bin', axis=1, inplace=True) # 一時的な列を削除

    # Structure Analysis: Correlation with Target Variable
    print_eda_section_header("構造分析: 目的変数との相関")
    correlations = df[numeric_cols].corr()[target_col].sort_values(ascending=False)
    purpose = "目的変数である中央住宅価格と各特徴量との線形関係の強さを評価する。"
    method = "ピアソン相関係数を計算し、目的変数との相関が高い特徴量を特定。"
    result = f"目的変数 '{target_col}' との相関:\n{correlations.to_string()}"
    interpretation = f"'median_income' は {target_col} と最も強い正の相関を示しており、所得が高い地域ほど住宅価格も高い傾向があることが示唆されます。"
    business_impact = "住宅価格予測モデルにおいて、特に重要度の高い特徴量を特定する。特徴量選択の根拠となる。"
    risk = "相関は因果関係ではないため、相関が高いからといってその特徴量だけが価格を決定するわけではない。多重共線性の可能性も考慮。"
    decision_impact = "High"
    hypothesis = "`median_income`が`median_house_value`の主要な予測因子である。"
    print_eda_output(purpose, method, result, interpretation, business_impact, risk, decision_impact, hypothesis)

    # Trend Analysis (Not Applicable for California Housing)
    print_eda_section_header("時系列分析: (適用外)")
    unnecessary_analyses.append("California Housingデータは時系列データではないため、トレンド分析は今回の意思決定に不要です。")

    print(f"\n**今回の意思決定に不要だった分析:**\n{'\n'.join(unnecessary_analyses) if unnecessary_analyses else '特になし。'}")
    print("\n" + "="*60 + "\n")

```

## 統計的検証レポート (Statistical Validation Report)

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats

# California Housingデータセットをロード
file_path = '/content/sample_data/california_housing_train.csv'
california_df = pd.read_csv(file_path)

print(f"--- {file_path} の統計的検証 ---")
```

### 仮説5の統計的検証: `median_income`が`median_house_value`の主要な予測因子である。

**検証方法**: `median_income`を独立変数、`median_house_value`を従属変数とする単回帰分析を実行し、`median_income`の係数の統計的有意性を評価します。

```python
# 回帰モデルの構築
# 目的変数: median_house_value, 説明変数: median_income
model_5 = ols('median_house_value ~ median_income', data=california_df).fit()

# モデルの要約を出力
print("### 仮説5の検証結果: median_income と median_house_value の単回帰分析 ###")
print(model_5.summary())

# 結果の解釈
print("\n--- 解釈 ---")
if model_5.pvalues['median_income'] < 0.05:
    print(f"事実（データ）: median_income のp値は {model_5.pvalues['median_income']:.3f} であり、0.05よりも小さいです。また、決定係数 (R-squared) は {model_5.rsquared:.3f} です。")
    print(f"構造（なぜそうなるか）: これは、統計的にmedian_incomeがmedian_house_valueに対して非常に強い予測力を持つことを示します。median_incomeの増加はmedian_house_valueの増加と関連しており、その関係性は統計的に有意であると言えます。決定係数の値から、median_incomeがmedian_house_valueの変動の約 {model_5.rsquared:.1%} を説明できることが示されます。")
    print("意味（ビジネス影響）: median_incomeは住宅価格の主要な決定要因であり、所得水準が高い地域ほど住宅価格も高くなるというビジネス的示唆を裏付けます。住宅市場の分析や投資戦略において、所得情報は非常に重要な要素となります。")
    print("リスク: 単回帰モデルであるため、median_income以外の要因が住宅価格に与える影響を考慮していません。また、相関は因果関係を意味しないため、このモデルだけで全てを説明できるわけではありません。")
else:
    print("事実（データ）: median_income のp値は0.05以上であり、統計的に有意な関係は見られませんでした。")
    print("構造（なぜそうなるか）: median_income単独ではmedian_house_valueを説明する十分な予測力がない可能性があります。")
    print("意味（ビジネス影響）: 住宅価格予測において、median_incomeのみに依存するのは危険であり、他の特徴量を組み合わせるか、より複雑なモデルを検討する必要があります。")
    print("リスク: 不十分なモデルに基づいた意思決定は誤った結論を導く可能性があります。")
```

### 仮説6の統計的検証: 築年数が若いほど、中央住宅価格は高い傾向がある。

**検証方法**: `housing_median_age`を区分（ビン）に分け、各区分における`median_house_value`の平均値に統計的に有意な差があるかを一元配置分散分析（ANOVA）を用いて検証します。

```python
# EDAで用いた築年数区分を再作成
california_df['age_bin'] = pd.cut(california_df['housing_median_age'],
                                  bins=[0, 10, 20, 30, 40, 50, california_df['housing_median_age'].max() + 1],
                                  labels=['0-10', '11-20', '21-30', '31-40', '41-50', '50+'],
                                  right=False)

# 築年数区分ごとの平均住宅価格を計算
mean_house_value_by_age = california_df.groupby('age_bin')['median_house_value'].mean().reset_index()

# ANOVAの実行
# statsmodelsのolsを使用してモデルを定義し、ANOVAテーブルを出力
# age_binがカテゴリカル変数であることを明示するためにC()を使用
model_6 = ols('median_house_value ~ C(age_bin)', data=california_df).fit()
anova_table = sm.stats.anova_lm(model_6, typ=2)

print("### 仮説6の検証結果: 築年数区分別 median_house_value のANOVA ###")
print("\n--- 築年数区分ごとの中央住宅価格（平均）---")
print(mean_house_value_by_age.to_string())
print("\n--- ANOVA検定結果 ---")
print(anova_table)

# 結果の解釈
print("\n--- 解釈 ---")
p_value_age_bin = anova_table['PR(>F)'][0] # C(age_bin)に対応するp値

if p_value_age_bin < 0.05:
    print(f"事実（データ）: 築年数区分（age_bin）のp値は {p_value_age_bin:.3f} であり、0.05よりも小さいです。これは、築年数区分によってmedian_house_valueの平均値に統計的に有意な差があることを示唆します。また、築年数区分ごとの平均値を見ると、特に'50+'の区分でmedian_house_valueが高くなる傾向がEDAの結果と同様に確認できます。")
    print("構造（なぜそうなるか）: 築年数という要因が住宅の価値に影響を与えているため、各区分間で住宅価格の分布に違いが生じています。特に、最も古い区分で高い平均値を示すことは、歴史的価値や再開発の可能性といった複雑な要因が絡んでいる可能性があります。これは「築年数が若いほど高い」という単純な仮説とは異なるパターンを示しています。")
    print("意味（ビジネス影響）: 住宅市場のセグメンテーションや価格戦略において、築年数は考慮すべき重要な要素です。ただし、必ずしも「若いほど高い」という直線的な関係ではなく、特定の築年数帯（例：50年以上）に価値が高まる現象があることを理解することが重要です。")
    print("リスク: ANOVAは全体としての差を検出しますが、どの区分間に具体的な差があるかまでは特定できません。より詳細な分析（例：多重比較検定）が必要になる場合があります。また、築年数だけでなく、他の多くの要因（リフォーム、立地、地域開発など）が住宅価格に影響を与えている可能性があります。")
else:
    print(f"事実（データ）: 築年数区分（age_bin）のp値は {p_value_age_bin:.3f} であり、0.05以上です。これは、築年数区分によってmedian_house_valueの平均値に統計的に有意な差があるとは言えないことを示唆します。")
    print("構造（なぜそうなるか）: 築年数だけでは住宅価格の変動を説明するのに不十分であるか、または築年数と住宅価格の間に単純な平均値の差が存在しないことを意味します。")
    print("意味（ビジネス影響）: 住宅価格予測において、築年数を主要な予測因子として用いる際には注意が必要であり、他の特徴量との組み合わせや、より詳細な分析が求められます。")
    print("リスク: 築年数のみに基づく住宅価格の評価や意思決定は、市場の現実を反映しない可能性があります。")

# 一時的な列を削除
california_df.drop('age_bin', axis=1, inplace=True)
```

*   **欠損値**: なし。
*   **重複行**: なし。

*   **分析可能な粒度**: カリフォルニア州の特定の地域（ブロック）における住宅に関する統計データの一行。

## 分析手法

1.  **Data Understanding (データ理解)**: 各データセットの行数・列数、データ型、欠損率、重複行数、主キー候補、分析可能な粒度を調査。
2.  **Data Cleaning (データクレンジング)**: 欠損値処理（今回は該当なし）、外れ値（IQR法）、業務ロジックに基づく異常値チェックを実施。California Housingデータでは多数の外れ値と異常値（例: `households` > `population`）を検出。
3.  **Adaptive EDA (探索的データ分析)**: データ構造に応じて分析項目を自動選択。分布分析、グループ比較分析、構造分析（相関分析、寄与度分析）を実施。時系列データではないためトレンド分析は適用外。
    *   **MNIST**: ラベル分布、ピクセル値統計、ラベルごとの平均ピクセル値、分散の大きいピクセルを分析。
    *   **California Housing**: `median_house_value`の分布、主要特徴量（`median_income`, `housing_median_age`）の分布、築年数区分別の中央住宅価格、目的変数との相関を分析。
4.  **Statistical Validation (統計的検証)**: EDAで導出された仮説の中から、ユーザー指定の仮説を統計的に検証。
    *   **仮説5**: `median_income`が`median_house_value`の主要な予測因子である。単回帰分析（OLS）で検証。
    *   **仮説6**: 築年数が若いほど、中央住宅価格は高い傾向がある。一元配置分散分析（ANOVA）で検証。
5.  **Insight Generation (インサイト生成)**: 統計的検証の結果に基づき、「事実（データ）」「構造（なぜそうなるか）」「意味（ビジネス影響）」「リスク」の形式で構造化されたインサイトを生成。
6.  **Recommendation (施策提言)**: インサイトに基づき、Growth Lever（増収）およびRisk Lever（構造リスク）に分類された具体的な施策を提案。各施策には、根拠EDA番号、期待効果、実行難易度、成功判定KPIを付与。
7.  **Impact × Effort Analysis (インパクト×エフォート分析)**: 提案された施策について、Impact (High/Medium/Low) と Effort (High/Medium/Low) を評価し、実行優先順位を決定。

## 主要インサイト

### インサイト1: 所得水準が住宅価格に強く影響する
*   **事実（データ）**: `median_income`は`median_house_value`の変動の約47.9%を説明し、統計的に非常に有意な予測因子です (p値 < 0.001)。
*   **構造（なぜそうなるか）**: 高所得層が多く住む地域では、住宅に対する購買力が高く、需要と供給のバランスが住宅価格を押し上げる傾向にあります。
*   **意味（ビジネス影響）**: 不動産開発や投資において、地域の中央所得は最も信頼性の高い指標の一つであり、高所得地域をターゲットとすることで高い収益性が期待できます。
*   **リスク**: 所得以外の要因を考慮しないと、予測の精度が低下する可能性があります。

### インサイト2: 築年数と住宅価格の関係は複雑であり、単純な線形傾向ではない
*   **事実（データ）**: 築年数区分によって`median_house_value`の平均値に統計的に有意な差がありますが (p値 < 0.001)、最も高い平均価格を示すのは「築年数50年以上」の区分でした。これは「築年数が若いほど高い」という単純な仮説とは異なります。
*   **構造（なぜそうなるか）**: 築浅物件のプレミアムに加え、築古物件の中には歴史的価値、再開発ポテンシャル、希少な立地条件などにより高い価値を持つものがあるためです。
*   **意味（ビジネス影響）**: 不動産市場は築年数だけで一律に評価すべきではなく、「築浅プレミアム」と「築古レガシー」といった異なるセグメントを認識した多様な戦略が必要です。
*   **リスク**: 築年数のみに注目した評価は、市場の現実を反映せず、誤った投資判断につながる可能性があります。

## 施策

### 1. 高所得地域をターゲットとした住宅開発・投資戦略の最適化
*   **根拠EDA番号**: California Housing EDA - 構造分析: 目的変数との相関（仮説5の統計的検証）
*   **期待効果**: `median_income`が高い地域に不動産開発や投資を集中させることで、高い販売価格や賃料収入が見込め、収益性を向上させることが期待されます。
*   **実行難易度**: Medium
*   **成功判定KPI**: ターゲット地域における新規開発物件の平均販売価格、投資案件の平均ROI、顧客層の平均`median_income`。

### 2. 築年数別セグメントに応じたマーケティング戦略の策定
*   **根拠EDA番号**: California Housing EDA - 比較分析: 築年数区分別の中央住宅価格（仮説6の統計的検証）
*   **期待効果**: 各築年数セグメントの特性と顧客ニーズに合致したマーケティングメッセージやチャネル戦略を展開することで、販売機会を最大化し、顧客エンゲージメントを向上させることが期待されます。
*   **実行難易度**: Medium
*   **成功判定KPI**: 各築年数セグメントにおける広告反応率、販売成約率、顧客満足度調査結果。

### 3. 築古高価格帯物件に対する詳細なデューデリジェンスプロセスの導入
*   **根拠EDA番号**: California Housing EDA - 比較分析: 築年数区分別の中央住宅価格（仮説6の統計的検証）
*   **期待効果**: 築年数50年以上の高価格帯物件に潜在するリスクを事前に特定し、予期せぬコストや投資損失を最小限に抑えることで、長期的な投資の安定性を確保します。
*   **実行難易度**: High
*   **成功判定KPI**: 築古物件取得後の予期せぬ大規模修繕費用の発生率、法的紛争発生件数、デューデリジェンスレポートにおけるリスク指摘事項の平均件数。

## 限界

*   **データセットの制約**: 分析に使用したデータセットは、特定の時点、特定の地域（カリフォルニア州）に限定されています。この結果が全ての地域や時点に普遍的に適用されるとは限りません。
*   **相関と因果関係**: 統計的検証で相関関係が確認された場合でも、それが直接的な因果関係を示すものではありません。他の潜在的な交絡因子が存在する可能性があります。
*   **モデルの単純性**: 仮説検証には単回帰分析やANOVAを使用しており、データに存在する複雑な非線形関係や多変量関係を完全に捉えきれていない可能性があります。
*   **ビジネス理解の初期段階**: 現時点ではビジネス課題が具体的に指定されていないため、一般的なビジネス示唆に留まっています。具体的な課題が定義されれば、より深い分析が可能です。
*   **外れ値処理**: IQR法による外れ値検出は実施しましたが、これらの外れ値の具体的な処理（除去、変換など）は行っていません。モデル構築時にはこれらの影響を考慮する必要があります。
